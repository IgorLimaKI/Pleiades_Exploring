```python
import pandas as pd
import re
import numpy as np
import json
```

```python
pl = pd.read_csv('PLEAIDES.csv')
```

```python
base = '{"nodes":[],"links":[]}'
```

```python
conj = json.loads(base)
```

```python
def makeNodes(jsn,data):
    for x in range(len(data)):
        aux = {"id":int(data.loc[x].id)}
        jsn["nodes"].append(aux)
        
    print('DONE')
    return jsn
```

```python
conj2 = makeNodes(conj, pl)
```

```python
def makeLinks(jsn, data):
    for i in range(len(data)):
        idd = int(data.at[i, 'id'])
        
        cw = str(data.at[i, 'connectsWith'])
        if(cw != "nan"):
            cw = cw.split(",")
            for x in cw:
                auc = {"source": idd,"target": int(float(x))}
                if(not(auc in jsn["links"])):
                    jsn["links"].append(auc)
        
        hc = str(data.at[i, 'hasConnectionsWith'])
        if(hc != "nan"):
            hc = hc.split(",")
            for y in hc:
                auh = {"source": int(float((y))),"target": idd}
                if(not(auh in jsn["links"])):
                    jsn["links"].append(auh)
    print('DONE')
    return jsn
```

```python
conj3 = makeLinks(conj2, pl)
```

```python
print(conj3["links"])
```

```python
with open('PLEIADES.json', 'w', encoding='utf-8') as f:
    json.dump(conj3, f, ensure_ascii=False, indent=4)
```
